```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel("average_input.xlsx")
target = pd.read_excel("input_data.xlsx",sheet_name='target')
```

```python

ax=data.plot(x='date', y='mian_stem_discharge(fts)',figsize=(20,5))
data.plot(x='date', y='Canaseraga_discharge(fts)',ax=ax)
data.plot(x='date', y='Oatka_discharge(fts)',ax=ax)
data.plot(x='date', y='Honeoye_discharge(fts)',ax=ax)
data.plot(x='date', y='Black_Creeks_dischage(fts)',ax=ax)
data.plot(x='date', y='Upper_Genesee_River_discharge(fts))',ax=ax)
ax.set_xlabel('')
ax.tick_params(axis="y", labelsize=20)
ax.tick_params(axis="x", labelsize=20)
ax.set_ylabel('Discharge(fts)',fontsize = 20)


ax2=data.plot(x='date', y='Canaseraga_TP_conc(mg/l)',figsize=(20,5))
data.plot(x='date', y='Oatka_TP_conc(mg/l)',ax=ax2)
data.plot(x='date', y='Honeoye_TP_conc(mg/l)',ax=ax2)
data.plot(x='date', y='Black_Creeks_TP_conc(mg/l)',ax=ax2)
data.plot(x='date', y='Upper_Genesee_River_TP_conc(mg/l)',ax=ax2)
ax2.set_ylim([0,5])
ax2.set_xlabel('')
ax2.tick_params(axis="y", labelsize=20)
ax2.tick_params(axis="x", labelsize=20)
ax2.set_ylabel('TP concentration(mg/l)',fontsize = 20)


ax3=data.plot(x='date', y='temp_max_average',figsize=(20,5))
data.plot(x='date', y='temp_min_average',ax=ax3)
data.plot(x='date', y='water_temp(main)',ax=ax3)
ax3.tick_params(axis="y", labelsize=20)
ax3.tick_params(axis="x", labelsize=20)
ax3.set_xlabel('Date',fontsize = 20)
ax3.set_ylabel('Temperature (degree)',fontsize = 20)
ax3.set_ylim([-20,40])
ax3.legend(loc='upper left')


ax4 = ax3.twinx()
data.plot(x='date', y='precipitation_average',figsize=(20,5),color='purple', ax=ax4)
ax4.set_ylabel('Precipitation(mm)',fontsize = 20)
ax4.tick_params(axis="y", labelsize=20)
```
